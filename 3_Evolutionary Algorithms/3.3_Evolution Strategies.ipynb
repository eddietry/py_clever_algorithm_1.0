```python
import math
import random
import numpy as np
```

```python
def get_obj(sol):
    return round(sum([ x**2 for x in sol]),2)

def get_sol(search_space):
    sol = []
    for i in range(len(search_space)):
        sol.append(round(random.uniform(search_space[i][0],search_space[i][1]),4))
    return sol

def get_best(pop):
    best = {}
    best['x'] = pop[0][0]
    best['fitness'] =  get_obj(best['x'])
    
    return best 

def init_population(search_space, pop_size):
    strategy = [ (sp[1]-sp[0])/20 for sp in search_space]
    pop = [ get_sol(search_space) for p in range(pop_size)]
    pop = [ (p,strategy) for p in pop]
    return pop


def mutate_problem(sol,strategy,minmax):
    new_x = sol 
    for i,v in enumerate(sol):
        new_x[i] = v + strategy[i] * np.random.randn()
        new_x[i] = max(new_x[i],minmax[i][0])
        new_x[i] = min(new_x[i],minmax[i][1])
    return new_x
def mutate_strategy(stdevs):
    tau = math.sqrt(2.0*len(stdevs))**-1.0
    tau_p = math.sqrt(2.0* math.sqrt(len(stdevs)))**-1.0
    child = [round(stdevs[i]*math.exp( tau_p * np.random.randn()+ tau * np.random.randn()),4) for i in range(len(stdevs))]
    return child
def get_mutate(sol,strategy,search_space):
    new_x = mutate_problem(sol,strategy,search_space)
    std = mutate_strategy(strategy)
    
    return new_x,std 

```

```python
pop = init_population(search_space,100)
pop.sort(key = lambda k: get_obj(k[0]))
best = get_best(pop)
print(best)
child_num = 30
n = 0 
while n < 1000:
    pop_next = [ get_mutate(p[0],p[1],search_space) for p in pop[:child_num]]
    pop_union = pop+pop_next
    pop_union.sort(key = lambda k: get_obj(k[0]))
    best_new = get_best(pop_union)
    pop = pop_union[:pop_size]
    if best_new['fitness'] < best['fitness']:
        best = best_new
```
