```python
import math
import random
import numpy as np
```

```python
def get_obj(sol):
    return round(sum([ x**2 for x in sol]),2)

def get_sol(search_space):
    sol = []
    for i in range(len(search_space)):
        sol.append(round(random.uniform(search_space[i][0],search_space[i][1]),3))
    return sol

def get_best(pop):
    best = {}
    best['x'] = pop[0]
    best['fitness'] =  get_obj(best['x'])
    return best
```

```python
def de_rand_1_bin(p0, p1, p2, p3, f, cr, search_space):
    sol = []
    cut  = random.sample(range(len(p0)),1)[0]
    for i in range(len(p0)):
        v = p0[i]
        if (i==cut or random.random() < cr):
            v = round(p3[i] + f * (p1[i] - p2[i]),3)
            if v < search_space[i][0]:
                v = search_space[i][0]
            if v > search_space[i][1]:
                v = search_space[i][1] 
        sol.append(v)   
    return sol 

def get_selection(pop, p0):
    selected = random.sample([ i for i in pop if i!= p0],3)
    return selected
```

```python
def get_next_pop(pop,search_space,weightf,crossf):
    pop_next = []
    for p0 in pop:
        p1,p2,p3 = get_selection(pop,p0)
        tmp = de_rand_1_bin(p0,p1,p2,p3,weightf,crossf,search_space)
        pop_next.append(tmp)
    return pop_next
```

```python
def select_population(pop,pop_next):
    tmp = []
    for i,p in enumerate(pop):
        if get_obj(p)< get_obj(pop_next[i]):
            tmp.append(p)
        else:
            tmp.append(pop_next[i])
    return tmp
```

```python
def search(max_gens, search_space, pop_size, f, cr):
    # get_pop
    pop = [ get_sol(search_space) for i in range(pop_size)]
    pop.sort(key = lambda  k:get_obj(k))
    # get_best
    best = get_best(pop)
    n = 0
    while n < max_gens:
        # get_next_pop()
        pop_next = get_next_pop(pop, search_space, f, cr)
        pop_next.sort(key = lambda  k:get_obj(k))
        # union_select()
        pop_new = select_population(pop, pop_next)
        pop_new.sort(key = lambda  k:get_obj(k))
        best_new = pop_new[0]
        if get_obj(best_new) < best['fitness']:
            best['x'] = best_new
            best['fitness'] = get_obj(best_new)
        print(best)
        n += 1
    
    return best
```

```python
# problem configuration
problem_size = 3
search_space = [ [-5,5] for i in range(problem_size)]
# algorithm configuration
max_gens = 200
pop_size = 500
weightf = 0.8
crossf = 0.9
# execute the algorithm
# best = search(max_gens, search_space, pop_size, weightf, crossf)
```

```python
search(max_gens, search_space, pop_size, weightf, crossf)
```
