```python
# 利用Harmony Search最小化cost目标函数
```

```python
import math
import random
```

```python
def get_obj(sol):
    return round(sum([ x**2 for x in sol]),2)
```

```python
def get_sol(search_space):
    sol = []
    for i in range(len(search_space)):
        sol.append(random.uniform(search_space[i][0],search_space[i][1]))
    return sol
```

```python
def get_init_harmony_memory(search_space, mem_size, factor=3):
    memory = [ get_sol(search_space) for i in range(mem_size*factor)]
    memory.sort(key=lambda k: get_obj(k))
    return memory
```

```python
def get_new_x(bounds, memory, consid_rate, adjust_rate, range_):
    new_x = [1,2,4]
    for i in  range(len(bounds)):
        if random.random() < adjust_rate:
            tmp = random.choice(memory)[i]
            if random.random() < consid_rate:
                tmp = tmp + range_ * random.uniform(-1,1)
                if tmp < bounds[i][0]:
                    tmp = bounds[i][0]
                if tmp > bounds[i][1]:
                    tmp = bounds[i][1]
                new_x[i] = tmp
        else:
            new_x[i] = random.uniform(bounds[i][0],bounds[i][1])
    return new_x
```

```python
def search(bounds, max_iter, mem_size, consid_rate, adjust_rate, range_):
    memory = get_init_harmony_memory(bounds, mem_size)
    print(type(memory))
    best = memory[0]
    for it in range(max_iter):
        new_x = get_new_x(bounds, memory, consid_rate, adjust_rate, range_)
        if  get_obj(new_x)< get_obj(best):
            print(get_obj(new_x))
            print(get_obj(best))
            memory.append(new_x)
            memory.sort(key=lambda k: get_obj(k))
            memory.pop()
            print("iters_{}".format(it))
            print("Best_fitness is {}".format(get_obj(memory[0])))
    
    return new_x
        
```

```python
# 问题相关的参数
problem_size = 3
bounds = [ [-5,5] for i in range(problem_size)]
# 算法相关参数
range_ = 0.02
max_iter = 5000
mem_size = 20
consid_rate = 0.95
adjust_rate = 0.7
```

```python
search(bounds, 10000, mem_size, consid_rate, adjust_rate, range_)
```
