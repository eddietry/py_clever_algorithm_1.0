```python
import math
import random
import numpy as np
```

```python
def get_obj(sol):
    return round(sum([ x**2 for x in sol]),2)
        
def get_sol(search_space):
    sol = []
    for i in range(len(search_space)):
        sol.append(round(random.uniform(search_space[i][0],search_space[i][1]),4))

    return sol
```

```python
def create_neigh_bee(site,patch_size,search_space):
    vector = []
    for i,v in enumerate(site):
        if random.random() < 0.5:
            v = v + random.random()* patch_size
        else:
            v = v - random.random()* patch_size
        if v < search_space[i][0]:
            v = search_space[i][0]
        if v > search_space[i][1]:
            v = search_space[i][1] 
        vector.append(v)
   
    return vector
```

```python
def search_neigh(parent, neigh_size, patch_size, search_space):
    neigh = []
    for i in range(neigh_size):
        neigh.append(create_neigh_bee(parent,patch_size,search_space))
    neigh.sort(key = lambda k:get_obj(k))
    return neigh              
```

```python
def create_scout_bees(search_space,num_scouts):
    return [ get_sol(search_space) for i in range(num_scouts)]
```

```python
def search(max_gens, search_space, num_bees, num_sites, elite_sites, patch_size, e_bees, o_bees):
    pop = [ get_sol(search_space) for i in range(num_bees)]
    pop.sort(key = lambda k:get_obj(k))
    best={'vector':pop[0]}
    best['fitness'] = get_obj(pop[0])
    n = 0 
    while n < max_gens:
        next_gen = []
        for i,parent in enumerate(pop[:num_sites]):
            if i < elite_sites:
                neigh_size = e_bees
            else:
                neigh_size = o_bees
            tmp= search_neigh(parent,neigh_size,patch_size,search_space)
            next_gen.append(tmp)
#         print(next_gen)
        next_gen = sum(next_gen,[])
#         print(next_gen)
        scouts = create_scout_bees(search_space, (num_bees-num_sites))
        pop = next_gen+scouts
        print(pop[0])
        pop.sort(key = lambda k:get_obj(k))
        
        if  get_obj(pop[0])< get_obj(best['vector']):
            best['vector'] = pop[0]
            best['fitness'] = get_obj(best['vector'])
        patch_size = patch_size * 0.95
        n += 1
    return best
```
